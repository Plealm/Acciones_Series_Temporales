```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parmetros
n = 1000
np.random.seed(42)  # Para reproducibilidad

# Generar procesos Gaussianos independientes
a1 = np.random.normal(0, 1, n)
a2 = np.random.normal(0, 1, n)

# Simular caminatas aleatorias
Z1 = np.zeros(n)
Z2 = np.zeros(n)

for t in range(1, n):
    Z1[t] = Z1[t-1] + a1[t]
    Z2[t] = Z2[t-1] + a2[t]

# Ajustar regresin
slope, intercept, r_value, p_value, std_err = stats.linregress(Z1, Z2)

# Visualizar resultados
plt.figure(figsize=(12, 6))
plt.plot(Z1, label='Z1')
plt.plot(Z2, label='Z2')
plt.title('Caminatas Aleatorias Z1 y Z2')
plt.legend()
plt.show()

plt.figure(figsize=(8, 8))
plt.scatter(Z1, Z2, alpha=0.5)
plt.plot(Z1, intercept + slope*Z1, color='red', label='Regresin')
plt.title('Regresin Z2 vs Z1')
plt.xlabel('Z1')
plt.ylabel('Z2')
plt.legend()
plt.show()

print(f"Pendiente: {slope:.4f}")
print(f"Intercepto: {intercept:.4f}")
print(f"R-cuadrado: {r_value**2:.4f}")
print(f"Valor p: {p_value:.4f}")
print(f"Error estndar: {std_err:.4f}")
```
